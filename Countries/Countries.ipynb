```python
def saveLatexTableToFile(df, filename="table.tex"):
    latex_str = "\\begin{table*}[h!]\n\\centering\n\\begin{tabular}{lllllll}\n\\hline\\hline\n"
    latex_str += "Country & NW Corner (Lat, Long) & SE Corner (Lat, Long) & Area($km^2$) & GDP(\\$) & Population & GDP per Capita(\\$) \\\\\n\\hline\n"
    
    for index, row in df.iterrows():
        nw_corner = f"({row['Upper Latitude']}, {row['Lower Longitude']})"
        se_corner = f"({row['Lower Latitude']}, {row['Upper Longitude']})"
        area = f"{row['Area']:.2e}"
        gdp = f"{row['GDP']:.2e}"
        population = f"{row['Population']:.2e}"
        gdp_per_capita = f"{row['GDP per Capita']:.2f}"
        
        latex_str += f"{index} & {nw_corner} & {se_corner} & {area} & {gdp} & {population} & {gdp_per_capita} \\\\\n"
    
    latex_str += "\\hline\\hline\\end{tabular}\n\\caption{Your caption here.}\n\\end{table*}"
    
    # Write the LaTeX string to a file
    with open(filename, 'w') as file:
        file.write(latex_str)
    print(f"LaTeX table saved to {filename}")

# Example usage:
saveLatexTableToFile(df, "CountriesTable.tex")
```
